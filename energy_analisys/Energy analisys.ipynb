```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
hERG1 = pd.read_csv("/home/nikita/D/Potassium channels/for_all/hERG1/energy_analisys/all_rtotal.csv")
hERG2 = pd.read_csv("/home/nikita/D/Potassium channels/for_all/hERG2/energy_analisys/all_rtotal.csv")
hERG3 = pd.read_csv("/home/nikita/D/Potassium channels/for_all/hERG3/energy_analisys/all_rtotal.csv")
rERG1 = pd.read_csv("/home/nikita/D/Potassium channels/for_all/rERG1/energy_analisys/all_rtotal.csv")
rERG2 = pd.read_csv("/home/nikita/D/Potassium channels/for_all/rERG2/energy_analisys/all_rtotal.csv")
rERG3 = pd.read_csv("/home/nikita/D/Potassium channels/for_all/rERG3/energy_analisys/all_rtotal.csv")
EAG = pd.read_csv("/home/nikita/D/Potassium channels/for_all/EAG/energy_analisys/all_rtotal.csv")
```

```python
bekm1_res_num = [i for i in range(1, 37)]
BeKm_1= ['ARG1','PRO2','THR3','ASP4','ILE5','LYS6','CYS7','SER8','GLU9','SER10','TYR11','GLN12','CYS13','PHE14','PRO15','VAL16','CYS17','LYS18','SER19','ARG20','PHE21','GLY22','LYS23','THR24','ASN25','GLY26','ARG27','CYS28','VAL29','ASN30','GLY31','PHE32','CYS33','ASP34','CYS35','PHE36']
```

```python
plotdata = pd.DataFrame({
    "hERG1":list(hERG1[' full']),
    "hERG2":list(hERG2[' full']),
    "hERG3":list(hERG3[' full']),
    "rERG1":list(rERG1[' full']),
    "rERG2":list(rERG2[' full']),
    "rERG3":list(rERG3[' full']),
    "EAG":list(EAG[' full'])
    }, 
    index=BeKm_1
)

plotdata.plot(kind="bar", figsize=(20,15))
plt.title("Energies by residue of ligand", fontsize=20, color='red')
plt.xlabel("BeKm-1 residues' numbers", fontsize=14)
plt.ylabel("Full energy", fontsize=14)

plt.show()
```

```python
plotdata = pd.DataFrame({
    "rERG2":list(rERG2[' full']),
    "rERG3":list(rERG3[' full'])
    }, 
    index=BeKm_1
)
ploterror = pd.DataFrame({
    "rERG2":list(rERG2[' stdfull']),
    "rERG3":list(rERG3[' stdfull'])
    }, 
    index=BeKm_1
)

plotdata.plot(kind="bar", figsize=(20,15), yerr=ploterror)
plt.title("rERG2 and rERG3 comparison", fontsize=20, color='red')
plt.xlabel("BeKm-1 residues' numbers", fontsize=14)
plt.ylabel("Full energy", fontsize=14)

plt.show()
```
