```python
def plot_dependance(data, labels, initial_levels, colors, title):
    fig = plt.figure(figsize=(12, 9))

    def plot_one_element_of_data(datum, label, initial_level, color):
        plt.plot(datum[columns[0]], datum[columns[1]], label=label, color=color)
        plt.plot(datum[columns[0]], np.ones(len(datum[columns[0]])) * initial_level, color=color, linestyle='--', label='Initial distance')
        
    for i in range(len(labels)):
        plot_one_element_of_data(data[i], labels[i], initial_levels[i], colors[i]) 
    
    plt.xlabel(columns[0], fontsize=14)
    plt.ylabel(columns[1], fontsize=14)
    plt.title(title, fontsize=20)
    plt.legend()

    #plt.savefig(title + ".png")
    plt.show()
```
