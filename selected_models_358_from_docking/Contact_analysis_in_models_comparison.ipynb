```python
import pandas as pd
```

```python
#Auxiliary lists of titles
columns1 = ['Molecular block number 1', 'Molecule number 1', 'Serial number of a residue 1', 
            'Title of a residue 1', 'PDB number of a residue 1']
columns2 = ['Molecular block number 2', 'Molecule number 2', 'Serial number of a residue 2', 
            'Title of a residue 2', 'PDB number of a residue 2']
necessary_columns = ['Title of a residue 1', 'PDB number of a residue 1',
              'Title of a residue 2', 'PDB number of a residue 2', 'Duration']

names = ['a', 'b', 'Duration']
durations = ['Duration 3', 'Duration 5', 'Duration 8']
```

```python
#Assign function which processes data from Nikolay csv file
def process_of_data(data):
    return pd.concat([pd.DataFrame(data.a.str.split().tolist(),
                                 columns = columns1), pd.DataFrame(data.b.str.split().tolist(),
                                 columns = columns2), pd.DataFrame(data.Duration)], axis=1)

#Assign filter function, extracting appropritate data from pandas DataFrame
def extract_appropriate_data(dataframe, duration_of_interaction = 0.1):
    return dataframe[(((pd.to_numeric(dataframe['PDB number of a residue 1']) < 1000) & 
                        (pd.to_numeric(dataframe['PDB number of a residue 2']) >= 1000)) | 
                       ((pd.to_numeric(dataframe['PDB number of a residue 1']) >= 1000) & 
                        (pd.to_numeric(dataframe['PDB number of a residue 2']) < 1000))) &
                    (pd.to_numeric(dataframe['Duration']) > duration_of_interaction)]
```

```python
#Model 3
data1 = pd.read_csv("./model_3/md.tpr_hb_ft.csv", names=names)
data2 = pd.read_csv("./model_3/md.tpr_phob_ft.csv", names=names)
data3 = pd.read_csv("./model_3/md.tpr_pi_ft.csv", names=names)
data4 = pd.read_csv("./model_3/md.tpr_sb_ft.csv", names=names)
data5 = pd.read_csv("./model_3/md.tpr_stk_ft.csv", names=names)

hb_3 = extract_appropriate_data(process_of_data(data1))[necessary_columns].rename(columns={'Duration': 'Duration 3'})
phob_3 = extract_appropriate_data(process_of_data(data2))[necessary_columns].rename(columns={'Duration': 'Duration 3'})
pi_3 = extract_appropriate_data(process_of_data(data3))[necessary_columns].rename(columns={'Duration': 'Duration 3'})
sb_3 = extract_appropriate_data(process_of_data(data4))[necessary_columns].rename(columns={'Duration': 'Duration 3'})
stk_3 = extract_appropriate_data(process_of_data(data5))[necessary_columns].rename(columns={'Duration': 'Duration 3'})
```

```python
data1 = pd.read_csv("./model_8/md.tpr_hb_ft.csv", names=names)
data2 = pd.read_csv("./model_8/md.tpr_phob_ft.csv", names=names)
data3 = pd.read_csv("./model_8/md.tpr_pi_ft.csv", names=names)
data4 = pd.read_csv("./model_8/md.tpr_sb_ft.csv", names=names)
data5 = pd.read_csv("./model_8/md.tpr_stk_ft.csv", names=names)

hb_8 = extract_appropriate_data(process_of_data(data1))[necessary_columns].rename(columns={'Duration': 'Duration 8'})
phob_8 = extract_appropriate_data(process_of_data(data2))[necessary_columns].rename(columns={'Duration': 'Duration 8'})
pi_8 = extract_appropriate_data(process_of_data(data3))[necessary_columns].rename(columns={'Duration': 'Duration 8'})
sb_8 = extract_appropriate_data(process_of_data(data4))[necessary_columns].rename(columns={'Duration': 'Duration 8'})
stk_8 = extract_appropriate_data(process_of_data(data5))[necessary_columns].rename(columns={'Duration': 'Duration 8'})
```

```python
def triple_left_join(all_residues, df1, df2, df3, columns = necessary_columns[:-1]):
    return all_residues.merge(df1.merge(df2.merge(df3, how='left', on=columns), how='left', on=columns), how='left', on=columns)
```

```python
def triple_union(df1, df2, df3):
    return pd.concat([df1, df2, df3])
```

```python
def final_processed_data(df1, df2, df3, columns = necessary_columns[:-1]):
    unique_pairs_of_residues = triple_union(df1[columns], 
                                            df2[columns], 
                                            df3[columns]).drop_duplicates()
    unique_pairs_of_residues = unique_pairs_of_residues.merge(df1, how='left', on=columns)
    unique_pairs_of_residues = unique_pairs_of_residues.merge(df2, how='left', on=columns)
    unique_pairs_of_residues = unique_pairs_of_residues.merge(df3, how='left', on=columns)
    return unique_pairs_of_residues
    
```

```python
#Hydrogen bonds
final_processed_data(hb_3, hb_5, hb_8).sort_values(by=durations, ascending=False)
```

```python
#Pi-cationic interaction
final_processed_data(pi_3, pi_5, pi_8).sort_values(by=durations, ascending=False)
```

```python
#Salt bridges
final_processed_data(sb_3, sb_5, sb_8).sort_values(by=durations, ascending=False)
```

```python
#Stacking
final_processed_data(stk_3, stk_5, stk_8).sort_values(by=durations, ascending=False)
```
