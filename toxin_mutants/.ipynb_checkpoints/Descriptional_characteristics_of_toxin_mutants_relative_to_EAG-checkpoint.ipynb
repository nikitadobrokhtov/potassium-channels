```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def RMS_from_xvg(path, skip_beg=18, skip_end=1):
    with open(path, "r") as myFile:
        data = [[float(i) for i in line.strip().split('    ')] for line in myFile.read().split('\n')[skip_beg:(skip_end * (-1))]]
    return [x[1] for x in data]
```

```python
def plot_xvg(path, label, data_size, skip_beg=18, skip_end=1):
    def RMS_from_xvg(path, skip_beg, skip_end):
        with open(path, "r") as myFile:
            data = [[float(i) for i in line.strip().split('    ')] for line in myFile.read().split('\n')[skip_beg:(skip_end * (-1))]]
        return [x[0] for x in data], [x[1] for x in data]
    x, y = RMS_from_xvg(path, skip_beg, skip_end)
    plt.plot(x[:data_size], y[:data_size], label=label)
```

```python
fig = plt.figure(figsize=(12,9))

#plot_xvg("/home/nikita/D/Potassium channels/for_all/EAG/pictures/RMSD_rec_rel_rec.xvg", label="Native", data_size=5001)
plot_xvg("./BeKm-1_ARG20/descriptional_characteristics/RMSD_tox_rel_tox_R20K.xvg", label="R20K", data_size=5001)
plot_xvg("./BeKm-1_ARG27/descriptional_characteristics/RMSD_tox_rel_tox_R27K.xvg", label="R27K", data_size=5001)
plot_xvg("./BeKm-1_amydated_LYS36/descriptional_characteristics/RMSD_tox_rel_tox_F36K.xvg", label="F36K", data_size=5001)
plot_xvg("./BeKm-1_PHE14-SER_ARG20-LYS_ARG27-LYS_PHE36-LYS_amydated/descriptional_characteristics/RMSD_tox_rel_tox_mul_14-20-27-36.xvg", label="Mul_14-20-27-36", data_size=5001)




plt.xlim(0, 100000)

plt.legend()
plt.xlabel("Time, ps", fontsize=14)
plt.ylabel("RMSD", fontsize=14)
plt.title("RMSD Ligand on Ligand", color='red', fontsize=20)

plt.show()
```

```python
def xvg_data(path, return_x = False, skip_beg=17, skip_end=1):
    def RMS_from_xvg(path, skip_beg, skip_end):
        with open(path, "r") as myFile:
            data = [[float(i) for i in line.strip().split('   ')] for line in myFile.read().split('\n')[skip_beg:(skip_end * (-1))]]
        return [x[0] for x in data], [x[1] for x in data]
    x, y = RMS_from_xvg(path, skip_beg, skip_end)
    if (return_x):
        return list(map(int, x)), y
    else:
        return y
```

```python
BeKm_1= ['ARG1','PRO2','THR3','ASP4','ILE5','LYS6','CYS7','SER8','GLU9','SER10','TYR11','GLN12','CYS13','PHE14','PRO15','VAL16','CYS17','LYS18','SER19','ARG20','PHE21','GLY22','LYS23','THR24','ASN25','GLY26','ARG27','CYS28','VAL29','ASN30','GLY31','PHE32','CYS33','ASP34','CYS35','PHE36','NH2']
```

```python
EAG_data = xvg_data("/home/nikita/D/Potassium channels/for_all/EAG/pictures/RMSF_tox.xvg")
EAG_data.append(0)
R20K = xvg_data("./BeKm-1_ARG20/descriptional_characteristics/RMSF_tox_R20K.xvg")
R20K.append(0)
R27K = xvg_data("./BeKm-1_ARG27/descriptional_characteristics/RMSF_tox_R27K.xvg")
R27K.append(0)
F36K = xvg_data("./BeKm-1_amydated_LYS36/descriptional_characteristics/RMSF_tox_F36K.xvg")
Mul_14_20_27_36 = xvg_data("./BeKm-1_PHE14-SER_ARG20-LYS_ARG27-LYS_PHE36-LYS_amydated/descriptional_characteristics/RMSF_tox_mul_14-20-27-36.xvg")
```

```python
plotdata = pd.DataFrame({
    "EAG":EAG_data,
    "R20K":R20K,
    "R27K":R27K,
    "F36K":F36K,
    "Mul_14-20-27-36":Mul_14_20_27_36
    }, 
    index=BeKm_1
)


plotdata.plot(kind="bar", figsize=(20,15))
plt.title("RMSF Ligand", fontsize=20, color='red')
plt.xlabel("BeKm-1's residue number", fontsize=14)
plt.ylabel("RMSD", fontsize=14)

plt.show()
```
