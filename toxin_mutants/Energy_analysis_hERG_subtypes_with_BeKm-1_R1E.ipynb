```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
hERG1 = pd.read_csv("/home/nikita/D/Potassium channels/for_all/hERG1/energy_analisys/all_rtotal.csv")
hERG2 = pd.read_csv("/home/nikita/D/Potassium channels/for_all/hERG2/energy_analisys/all_rtotal.csv")
hERG3 = pd.read_csv("/home/nikita/D/Potassium channels/for_all/hERG3/energy_analisys/all_rtotal.csv")

hERG1_R1E = pd.read_csv("./BeKm-1_R1E_subtypes/hERG1/all_rtotal.csv")
hERG2_R1E = pd.read_csv("./BeKm-1_R1E_subtypes/hERG2/all_rtotal.csv")
hERG3_R1E = pd.read_csv("./BeKm-1_R1E_subtypes/hERG3/all_rtotal.csv")
```

```python
bekm1_res_num = [i for i in range(1, 37)]
BeKm_1= ['ARG1','PRO2','THR3','ASP4','ILE5','LYS6','CYS7','SER8','GLU9','SER10','TYR11','GLN12','CYS13','PHE14','PRO15','VAL16','CYS17','LYS18','SER19','ARG20','PHE21','GLY22','LYS23','THR24','ASN25','GLY26','ARG27','CYS28','VAL29','ASN30','GLY31','PHE32','CYS33','ASP34','CYS35','PHE36']
```

```python
plotdata = pd.DataFrame({
    "hERG1":list(hERG1[' full']),
    "hERG1_R1E":list(hERG1_R1E[' full'])
    }, 
    index=BeKm_1
)
ploterror = pd.DataFrame({
    "hERG1":list(hERG1[' stdfull']),
    "hERG1_R1E":list(hERG1_R1E[' stdfull'])
    }, 
    index=BeKm_1
)

plotdata.plot(kind="bar", figsize=(20,15), yerr=ploterror)
plt.title("hERG1 and hERG1_R1E comparison", fontsize=20, color='red')
plt.xlabel("BeKm-1's residue number", fontsize=14)
plt.ylabel("Full energy", fontsize=14)

plt.show()
```
