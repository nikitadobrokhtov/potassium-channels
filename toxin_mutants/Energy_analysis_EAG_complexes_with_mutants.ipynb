```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
native = pd.read_csv("/home/nikita/D/Potassium channels/for_all/EAG/energy_analisys/all_rtotal.csv")
R20K = pd.read_csv("./BeKm-1_ARG20/energy_analysis/new_conformation/all_rtotal.csv")
R27K = pd.read_csv("./BeKm-1_ARG27/energy_analysis/new_conformation/all_rtotal.csv")
F36K = pd.read_csv("./BeKm-1_amydated_LYS36/energy_analysis/new_conformation/all_rtotal.csv")
Mul_14_20_27_36 = pd.read_csv("./BeKm-1_PHE14-SER_ARG20-LYS_ARG27-LYS_PHE36-LYS_amydated/energy_analysis/new_conformation/all_rtotal.csv")
```

```python
ddg_native = str(native.sum(axis=0)[0].round(0))
ddg_R20K = str(R20K.sum(axis=0)[0].round(0))
ddg_R27K = str(R27K.sum(axis=0)[0].round(0))
ddg_F36K = str(F36K.sum(axis=0)[0].round(0))
ddg_Mul_14_20_27_36 = str(Mul_14_20_27_36.sum(axis=0)[0].round(0))
```

```python
bekm1_res_num = [i for i in range(1, 38)]
BeKm_1= ['ARG1','PRO2','THR3','ASP4','ILE5','LYS6','CYS7','SER8','GLU9','SER10','TYR11','GLN12','CYS13','PHE14','PRO15','VAL16','CYS17','LYS18','SER19','ARG20','PHE21','GLY22','LYS23','THR24','ASN25','GLY26','ARG27','CYS28','VAL29','ASN30','GLY31','PHE32','CYS33','ASP34','CYS35','PHE36', 'NH2']
```

### Prepare arrays

```python
native_data = list(native[' full'])
native_data.append(0)
native_error = list(native[' stdfull'])
native_error.append(0)


R20K_data = list(R20K[' full'])
R20K_error = list(R20K[' stdfull'])

R27K_data = list(R27K[' full'])
R27K_error = list(R27K[' stdfull'])

F36K_data = list(F36K[' full'])
F36K_error = list(F36K[' stdfull'])

Mul_14_20_27_36_data = list(Mul_14_20_27_36[' full'])
Mul_14_20_27_36_error = list(Mul_14_20_27_36[' stdfull'])
```

# All

```python
plotdata = pd.DataFrame({
    "Native":native_data,
    "R20K":R20K_data,
    "R27K":R27K_data,
    "F36Kam":F36K_data,
    "F14S-R20K-R27K-F36Kam":Mul_14_20_27_36_data
    }, 
    index=BeKm_1
)

plotdata.plot(kind="bar", figsize=(20,15))
plt.title("Energies by residue of ligand", fontsize=20, color='red')
plt.xlabel("BeKm-1 residues' numbers", fontsize=14)
plt.ylabel("Full energy", fontsize=14)

plt.show()
```

# BeKm-1_ARG20-LYS

```python
plotdata = pd.DataFrame({
    "Native":native_data,
    "R20K":R20K_data
    }, 
    index=BeKm_1
)
ploterror = pd.DataFrame({
    "Native":native_error,
    "R20K":R20K_error
    }, 
    index=BeKm_1
)

plotdata.plot(kind="bar", figsize=(20,15), yerr=ploterror)
plt.title("ARG20 -> LYS\n \"$\Delta\Delta G_{nat} = $ " + 
          ddg_native + "\"\n \"$\Delta\Delta G_{R20K} = $ " + 
          ddg_R20K + "\"", fontsize=20, color='red')
plt.xlabel("BeKm-1 residues' numbers", fontsize=14)
plt.ylabel("Full energy", fontsize=14)
plt.show()
```
