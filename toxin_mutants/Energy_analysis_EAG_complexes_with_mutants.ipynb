```python
#plotdata.plot(kind="bar", figsize=(20,15), yerr=ploterror, fontsize=20, capsize=3)
def constcuct_bar(EAG_full, EAG_error, mut_label, hERG_full=native_data, hERG_error=native_error, BeKm_1=BeKm_1):
    fig = plt.figure(figsize=(20,15)) 

    x_axis = np.arange(len(BeKm_1))

#     hERG_full = list(hERG1[' full'])
#     EAG_full = list(EAG[' full'])

#     hERG_error = list(hERG1[' stdfull'])
#     EAG_error = list(EAG[' stdfull'])

    width = 0.4
    dx = 0.2

    auxiliary_error = np.zeros(37)

    plt.bar(x_axis - dx,hERG_full, width=width, label="Native")
    plt.bar(x_axis + dx,EAG_full, width=width, label=mut_label)

    def error_2d_data(data, error):
        min_ = []
        max_ = []
        for num, i in enumerate(error):
            if (data[num] <= 0):
                min_.append(i)
                max_.append(0)
            else:
                min_.append(0)
                max_.append(i)
        return [min_, max_]
    



    plt.errorbar(x_axis - dx,hERG_full,error_2d_data(hERG_full, hERG_error),ls="None", capsize=3)
    plt.errorbar(x_axis + dx,EAG_full,error_2d_data(EAG_full, EAG_error),ls="None", capsize=3)

    plt.xticks(x_axis, BeKm_1, rotation=90, fontsize=20)
    plt.ylabel("Full energy, kJ/mol", fontsize=20)
    plt.xlabel("BeKm-1's residue number", fontsize=20)
    plt.legend(fontsize=20)
    


```

```python
constcuct_bar(R20K_data, R20K_error, "R20K")
plt.title("ARG20 $\Longrightarrow$ LYS\n \"$\Delta G_{nat} = $ " + 
          ddg_native + " kJ/mol\"\n \"$\Delta G_{R20K} = $ " + 
          ddg_R20K + " kJ/mol\"", fontsize=25, color='red')

plt.savefig("model1.png")

plt.show()
```
