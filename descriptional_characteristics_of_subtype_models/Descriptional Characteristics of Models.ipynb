```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def RMS_from_xvg(path, skip_beg=18, skip_end=1):
    with open(path, "r") as myFile:
        data = [[float(i) for i in line.strip().split('    ')] for line in myFile.read().split('\n')[skip_beg:(skip_end * (-1))]]
    return [x[1] for x in data]
```

```python
def plot_xvg(path, label, data_size, skip_beg=18, skip_end=1):
    def RMS_from_xvg(path, skip_beg, skip_end):
        with open(path, "r") as myFile:
            data = [[float(i) for i in line.strip().split('    ')] for line in myFile.read().split('\n')[skip_beg:(skip_end * (-1))]]
        return [x[0] for x in data], [x[1] for x in data]
    x, y = RMS_from_xvg(path, skip_beg, skip_end)
    plt.plot(x[:data_size], y[:data_size], label=label)
```

```python
fig = plt.figure(figsize=(12,9))

plot_xvg("./hERG1/pictures/RMSD_rec_rel_rec.xvg", label="hERG1", data_size=1001)
plot_xvg("./hERG2/pictures/RMSD_rec_rel_rec.xvg", label="hERG2", data_size=1001)
plot_xvg("./hERG3/pictures/RMSD_rec_rel_rec.xvg", label="hERG3", data_size=1001)
plot_xvg("./rERG1/pictures/RMSD_rec_rel_rec.xvg", label="rERG1", data_size=1001)
plot_xvg("./rERG2/pictures/RMSD_rec_rel_rec.xvg", label="rERG2", data_size=1001)
plot_xvg("./rERG3/pictures/RMSD_rec_rel_rec.xvg", label="rERG3", data_size=1001)


plt.xlim(0, 100000)

plt.legend()
plt.xlabel("Time, ps", fontsize=14)
plt.ylabel("RMSD", fontsize=14)
plt.title("RMSD Receptor on Receptor", color='red', fontsize=20)

plt.show()
```

```python
def xvg_data(path, skip_beg=17, skip_end=1):
    def RMS_from_xvg(path, skip_beg, skip_end):
        with open(path, "r") as myFile:
            data = [[float(i) for i in line.strip().split('   ')] for line in myFile.read().split('\n')[skip_beg:(skip_end * (-1))]]
        return [x[0] for x in data], [x[1] for x in data]
    x, y = RMS_from_xvg(path, skip_beg, skip_end)
    return y
```

```python
fig = plt.figure(figsize=(12,9))

toxin_atoms=[i for i in range(1, 548)]

plt.plot(toxin_atoms, xvg_data("./hERG1/pictures/RMSF_tox.xvg"), label="hERG1")
plt.plot(toxin_atoms, xvg_data("./hERG2/pictures/RMSF_tox.xvg"), label="hERG2")
plt.plot(toxin_atoms, xvg_data("./hERG3/pictures/RMSF_tox.xvg"), label="hERG3")
plt.plot(toxin_atoms, xvg_data("./rERG1/pictures/RMSF_tox.xvg"), label="rERG1")
plt.plot(toxin_atoms, xvg_data("./rERG2/pictures/RMSF_tox.xvg"), label="rERG2")
plt.plot(toxin_atoms, xvg_data("./rERG3/pictures/RMSF_tox.xvg"), label="rERG3")



plt.legend()
plt.xlabel("Atoms in ligand", fontsize=14)
plt.ylabel("RMSD", fontsize=14)
plt.title("RMSF Ligand", color='red', fontsize=20)

plt.show()
```

```python
def data_for_average(label):
    d1 = np.array(xvg_data("./" + label + "/pictures/RMSF_rec_1.xvg"))
    d2 = np.array(xvg_data("./" + label + "/pictures/RMSF_rec_2.xvg"))
    d3 = np.array(xvg_data("./" + label + "/pictures/RMSF_rec_3.xvg"))
    d4 = np.array(xvg_data("./" + label + "/pictures/RMSF_rec_4.xvg"))
    
    if (label == "hERG2"):
        d4 = d4[:-1]
    if (label == "hERG1"):
        d3 = d3[:-1]
    
    d_aver = (d1 + d2 + d3 + d4) / 4
    d_std = np.std([d1, d2, d3, d4], axis=0)
    return d_aver, d_std

def plot_receptor_RMSF(label):
    fig = plt.figure(figsize=(12,9))

    receptor_name = label

    d_aver, d_std = data_for_average(receptor_name)

    plt.plot(d_aver - d_std, label="Low border", alpha=0.5)
    plt.plot(d_aver + d_std, label="High border", alpha=0.5)


    #plt.xlim(2000, 3000)

    plt.legend()
    plt.xlabel("Atoms in receptor subunit", fontsize=14)
    plt.ylabel("RMSD", fontsize=14)
    plt.title("RMSF " + receptor_name + " average", color='red', fontsize=20)

    plt.show()    
```

```python
plot_receptor_RMSF("hERG1")
```

```python
plot_receptor_RMSF("hERG2")
```

```python
plot_receptor_RMSF("hERG3")
```

```python
plot_receptor_RMSF("rERG1")
```

```python
plot_receptor_RMSF("rERG2")
```

```python
plot_receptor_RMSF("rERG3")
```
